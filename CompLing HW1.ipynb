```python
import re
```

```python
def FindHyphen(filename):
    dictionary={}
    for line in open(filename,encoding='utf-8'):
        for i in re.findall('\s[А-я]+\-[А-я]+\s',line):
            i=i.lower()
            i=i.strip()
            dictionary.setdefault(i,0)
            dictionary[i]+=1
    return dictionary    
```

```python
print(FindHyphen('Владислав Крапивин.txt'))
```

```python
def is_abbr(string):
    if re.match('[А-Я]{2,100}',string) or re.match('([А-Я]\.){2,100}',string)\
        or re.match('[цкнгшщзхфвпрлджчсмтбЦКНГШЩЗХФВПРЛДЖЧСМТБ]+\.?',string):
        return True
    else:
        return False
```

```python
def Find_Abbr(filename):
    abbr={}
    for line in open(filename,encoding='utf-8'):
        for i in re.findall('\s[А-я]+[-]*[А-я]*\s',line):
            if is_abbr(i):
                
                abbr.setdefault(i,0)
                abbr[i]+=1
    return abbr
```

```python
Find_Abbr('Владислав Крапивин.txt')
```
